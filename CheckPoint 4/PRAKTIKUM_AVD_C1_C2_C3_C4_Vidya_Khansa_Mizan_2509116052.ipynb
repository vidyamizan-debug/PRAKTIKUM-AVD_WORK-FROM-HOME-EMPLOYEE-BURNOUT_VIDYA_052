### **Outliers**

```python
results = []

cols = burn.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = burn[col].quantile(0.25)
  q3 = burn[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = burn[(burn[col] < lower_bound) | (burn[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(burn))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_burn = pd.DataFrame(results)
results_burn.set_index('Kolom', inplace=True)
results_df = results_burn.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_burn)
```

## **Data Construction**

**Alasan :**

Pada tahap data construction ini saya akan menambahkan satu kolom baru yaitu kolom work_load, kolom ini berguna untuk membantu kita dalam melihat apakah pekerja dengan jam kerja yang lebih tinggi memiliki kecenderungan untuk memiliki tingkat burnout yang lebih besar

```python
def categorize_workload(work_hours):
    if work_hours > 9:
        return 'High'
    elif work_hours >= 7:
        return 'Medium'
    else:
        return 'Low'

burn['work_load'] = burn['work_hours'].apply(categorize_workload)
```

### **Heatmap**

Berikut ini ialah Heatmap yang saya buat dan gunakan untuk melihat hubungan atau korelasi antara beberapa variabel, yaitu work_hours (jam kerja), sleep_hours (jam tidur), screen_time_hours (waktu penggunaan layar), dan burnout_score (tingkat burnout). Warna pada heatmap menunjukkan seberapa kuat hubungan antar variabel: semakin terang warnanya, semakin kuat hubungannya. Dari grafik terlihat bahwa jam kerja memiliki hubungan yang sangat kuat dengan waktu penggunaan layar (nilai korelasi sekitar 0.95). Sementara itu, hubungan antara jam tidur dengan variabel lain terlihat sangat lemah atau hampir tidak ada.

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=burn_clean[['work_hours', 'sleep_hours', 'screen_time_hours', 'burnout_score']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')
plt.title('Korelasi Jam Kerja, Tidur, dan Screentime terhadao Skor Burnout')
plt.show()
```

**Insight :**


* Jam kerja memiliki hubungan yang sangat kuat dengan waktu penggunaan layar.

* Jam tidur hampir tidak memiliki hubungan dengan jam kerja maupun screen time.

* Skor burnout hanya memiliki hubungan yang lemah dengan jam kerja dan screen time.


**Action :**


* Melakukan analisis lebih lanjut untuk memahami mengapa jam kerja sangat berkaitan dengan screen time.

* Menambahkan variabel lain (misalnya tingkat stres atau beban pekerjaan) untuk melihat faktor yang lebih memengaruhi burnout.
