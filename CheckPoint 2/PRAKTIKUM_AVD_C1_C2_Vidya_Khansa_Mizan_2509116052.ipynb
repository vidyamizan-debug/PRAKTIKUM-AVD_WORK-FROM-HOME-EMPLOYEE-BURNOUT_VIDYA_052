### **Outliers**

```python
results = []

cols = burn.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = burn[col].quantile(0.25)
  q3 = burn[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = burn[(burn[col] < lower_bound) | (burn[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(burn))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_burn = pd.DataFrame(results)
results_burn.set_index('Kolom', inplace=True)
results_df = results_burn.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_burn)
```

## **Eksplorasi Data (EDA)**

### **Comparison/Perbandingan**

```python
avg_burnout_by_day_type = burn.groupby('day_type')['burnout_score'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=avg_burnout_by_day_type.index, y=avg_burnout_by_day_type.values, palette='viridis', hue=avg_burnout_by_day_type.index, legend=False)
plt.title('Average Burnout Score by Day Type')
plt.xlabel('Day Type')
plt.ylabel('Average Burnout Score')
plt.xticks(rotation=45)
plt.show()
```

**Kesimpulan :**


*   Rata-rata burnout score lebih tinggi pada Weekday dibandingkan Weekend. Ini menunjukkan bahwa tingkat kelelahan atau stres cenderung meningkat saat hari kerja dibandingkan akhir pekan.





### **Composition/Komposisi**

```python
burnout_risk_counts = burn['burnout_risk'].value_counts()

plt.figure(figsize=(10, 8))
plt.pie(burnout_risk_counts, labels=burnout_risk_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Proporsi Karyawan Berdasarkan Risiko Burnout')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

**Kesimpulan :**


*   Mayoritas karyawan berada pada kategori risiko burnout rendah (84.8%), sementara risiko sedang 14.1% dan tinggi hanya 1.1%. Ini menunjukkan secara umum kondisi masih terkendali, tetapi tetap perlu perhatian pada kelompok risiko sedang dan tinggi agar tidak meningkat.



### **Distribution/Distribusi**

```python
plt.figure(figsize=(10, 6))
sns.histplot(burn['work_hours'], bins=20, kde=True)
plt.title('Distribusi Jam Kerja')
plt.xlabel('Jam Kerja')
plt.ylabel('Frekuensi')
plt.show()
```

**Kesimpulan :**

* Distribusi pada tabel menunjukkan bahwa distribusinya ini condong ke skewed left.

*   Distribusi jam kerja terlihat menyebar antara sekitar 3 hingga 12 jam, dengan konsentrasi terbesar berada di kisaran 4â€“6 jam dan 8â€“9 jam. Ini menunjukkan adanya dua pola jam kerja dominan, kemungkinan mencerminkan perbedaan beban kerja antar karyawan atau tipe hari kerja.



### **Relationship/Hubungan**

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=burn[['work_hours', 'burnout_score']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')
plt.title('Korelasi antara Jam Kerja dan Tingkat Stress')
plt.show()
```

**Kesimpulan :**


*   Nilai korelasi sebesar 0.12 menunjukkan bahwa hubungan antara jam kerja dan tingkat stres sangat lemah dan hampir tidak signifikan. Artinya, peningkatan jam kerja tidak secara langsung berkaitan kuat dengan kenaikan burnout dalam data ini.
