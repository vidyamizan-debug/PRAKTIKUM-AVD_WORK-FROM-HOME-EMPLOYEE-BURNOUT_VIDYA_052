# **Business Understanding**

## **Business Objective**

Dataset Work From Home Employee Burnout

Data ini merupakan sebuah kumpulan data yang memuat kondisi karyawan selama melakukan Work From Home atau yang disingkat sebagai WFH. Ini digunakan agar kita dapat memahami faktor apa saja yang memengaruhi tingkat burnout, seperti jam kerja dan istirahat, serta work life balance.

Dalam dataset ini terdapat 1800 baris dan 11 kolom

## **Assess Situation**

Burnout ini sering menjadi isu bagi para karyawan yang menerapkan situasi WFH di pola kerjanya. Adanya perubahan pola kerja ini dapat memengaruhi tingkat stres yang dimiliki karyawan. Oleh karena itu, perusahaan perlu untuk melakukan analisis seperti dataset ini untuk bisa memahami penyebab burnout yang dialami oleh karyawan mereka.
